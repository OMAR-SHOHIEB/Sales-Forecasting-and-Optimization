```python
import pandas as pd
```

```python
df=pd.read_csv('After_EDA.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['store_encoded']=encoder.fit_transform(df['store'])
df['country_encoded']=encoder.fit_transform(df['country'])
df['product_encoded']=encoder.fit_transform(df['product'])
df.head()
```

```python
new_df=df[['day','month','year','country_encoded','store_encoded','product_encoded','num_sold']]
for lag in range(1, 8):
    new_df[f'lag_{lag}'] = new_df['num_sold'].shift(lag)
new_df['rolling_mean_7'] = new_df['num_sold'].shift(1).rolling(window=7).mean()
new_df['rolling_std_7'] = new_df['num_sold'].shift(1).rolling(window=7).std()
new_df.dropna()
new_df.to_csv('ready_df.csv',index=False)
```
